```python
"""
Portfolio ESG Analyzer - Partie 3/6 : Classe principale et framework
"""

class ESGPortfolioAnalyzer:
    """Classe principale pour l'analyse ESG d'un portefeuille"""

    def __init__(self, portfolio_size=10):
        self.portfolio_size = portfolio_size
        self.companies = []
        self.esg_framework = self._load_esg_framework()
        self.results = pd.DataFrame()
        self.detailed_scores = pd.DataFrame()
        self.recommendations = []
        self.portfolio_data = None
        self.insights = {}

    def _load_esg_framework(self):
        """Charge le framework ESG bas√© sur les normes internationales"""
        return {
            'environmental': {
                'categories': {
                    'GHG_Reduction': {'weight': 0.25, 'standards': ['GRI 305', 'TCFD']},
                    'Water_Efficiency': {'weight': 0.20, 'standards': ['GRI 303']},
                    'Waste_Recycling': {'weight': 0.20, 'standards': ['GRI 306']},
                    'Renewable_Energy': {'weight': 0.25, 'standards': ['SASB EN']},
                    'Biodiversity_Protection': {'weight': 0.10, 'standards': ['CSRD']}
                },
                'standards': ['GRI 305', 'SASB EN', 'TCFD', 'CSRD'],
                'total_weight': 0.40
            },
            'social': {
                'categories': {
                    'Employee_Satisfaction': {'weight': 0.25, 'standards': ['ISO 26000']},
                    'Diversity_Index': {'weight': 0.20, 'standards': ['SDG 5', 'SDG 10']},
                    'Safety_Record': {'weight': 0.25, 'standards': ['GRI 403']},
                    'Community_Investment': {'weight': 0.15, 'standards': ['GRI 413']},
                    'Human_Rights': {'weight': 0.15, 'standards': ['UN Guiding Principles']}
                },
                'standards': ['UN SDGs 8,10', 'ISO 26000', 'GRI 400'],
                'total_weight': 0.30
            },
            'governance': {
                'categories': {
                    'Board_Independence': {'weight': 0.30, 'standards': ['OCDE']},
                    'Ethics_Score': {'weight': 0.25, 'standards': ['ISO 37001']},
                    'Transparency_Index': {'weight': 0.20, 'standards': ['GRI 205']},
                    'Shareholder_Rights': {'weight': 0.15, 'standards': ['OCDE']},
                    'Anti_Corruption': {'weight': 0.10, 'standards': ['ISO 37001']}
                },
                'standards': ['OCDE Principles', 'ISO 37001', 'GRI 205'],
                'total_weight': 0.30
            }
        }

print("‚úÖ Partie 3/6 : Classe et framework d√©finis!")
```

```python
"""
Portfolio ESG Analyzer - Partie 4/6 : M√©thodes de chargement
"""

def load_portfolio_data(self, excel_file_path):
    """Charge les donn√©es du portefeuille depuis Excel"""
    try:
        print("üì• Chargement des donn√©es Excel...")

        # V√©rifier si le fichier existe
        if not os.path.exists(excel_file_path):
            print(f"‚ùå Fichier non trouv√©: {excel_file_path}")
            print("üìù Veuillez t√©l√©verser votre fichier Excel")
            return False

        # Lire les noms des feuilles
        xls = pd.ExcelFile(excel_file_path)
        print(f"üìã Feuilles disponibles: {xls.sheet_names}")

        # Lecture des diff√©rentes feuilles
        company_info = pd.read_excel(excel_file_path, sheet_name='Companies')
        esg_scores = pd.read_excel(excel_file_path, sheet_name='ESG_Scores')
        financial_data = pd.read_excel(excel_file_path, sheet_name='Financials')

        print(f"üìä Feuille 'Companies': {company_info.shape[0]} entreprises, {company_info.shape[1]} colonnes")
        print(f"üìä Feuille 'ESG_Scores': {esg_scores.shape[0]} entr√©es, {esg_scores.shape[1]} colonnes")
        print(f"üìä Feuille 'Financials': {financial_data.shape[0]} entr√©es, {financial_data.shape[1]} colonnes")

        # Afficher les colonnes disponibles
        print("\nüìã Colonnes dans 'Companies':", list(company_info.columns))
        print("üìã Colonnes dans 'ESG_Scores':", list(esg_scores.columns)[:10], "...")
        print("üìã Colonnes dans 'Financials':", list(financial_data.columns))

        # Fusion des donn√©es
        portfolio_data = pd.merge(company_info, esg_scores, on='Company_ID', how='left')
        portfolio_data = pd.merge(portfolio_data, financial_data, on='Company_ID', how='left')

        self.portfolio_data = portfolio_data
        print(f"\n‚úÖ Donn√©es charg√©es avec succ√®s pour {len(portfolio_data)} entreprises")

        # Afficher un aper√ßu
        print("\nüìã APER√áU DES DONN√âES (3 premi√®res entreprises):")
        print(portfolio_data[['Company_ID', 'Company_Name', 'Sector']].head(3))

        return True

    except Exception as e:
        print(f"‚ùå Erreur lors du chargement: {e}")
        print("\nüîç D√©pannage:")
        print("1. V√©rifiez que le fichier s'appelle 'portfolio_data.xlsx'")
        print("2. V√©rifiez qu'il contient 3 feuilles: Companies, ESG_Scores, Financials")
        print("3. V√©rifiez que chaque feuille a une colonne 'Company_ID'")
        return False

# Ajouter la m√©thode √† la classe
ESGPortfolioAnalyzer.load_portfolio_data = load_portfolio_data

print("‚úÖ Partie 4/6 : M√©thodes de chargement ajout√©es!")
```

```python
"""
Portfolio ESG Analyzer - Partie 5/6 : M√©thodes de calcul
"""

def calculate_detailed_scores(self):
    """Calcule les scores d√©taill√©s pour chaque crit√®re ESG"""
    if self.portfolio_data is None:
        raise ValueError("‚ùå Veuillez d'abord charger les donn√©es du portefeuille")

    print("üßÆ Calcul des scores d√©taill√©s...")
    detailed_results = []

    for _, company in self.portfolio_data.iterrows():
        company_scores = {
            'Company_ID': company['Company_ID'],
            'Company_Name': company['Company_Name']
        }

        # Calcul des scores par crit√®re
        for category, data in self.esg_framework.items():
            for criterion, weights in data['categories'].items():
                if criterion in company:
                    raw_score = company[criterion]
                    # Normalisation du score (0-100)
                    normalized_score = min(max(raw_score, 0), 100)
                    company_scores[f"{criterion}_Score"] = normalized_score
                    company_scores[f"{criterion}_Weight"] = weights['weight']

        detailed_results.append(company_scores)

    self.detailed_scores = pd.DataFrame(detailed_results)
    print(f"‚úÖ Scores d√©taill√©s calcul√©s pour {len(detailed_results)} entreprises")
    return self.detailed_scores

def _calculate_weighted_score(self, company, category):
    """Calcule le score pond√©r√© pour une cat√©gorie sp√©cifique"""
    total_weight = 0
    weighted_sum = 0

    for criterion, weights in self.esg_framework[category]['categories'].items():
        if criterion in company:
            raw_score = company[criterion]
            # Assurer que le score est entre 0 et 100
            normalized_score = min(max(raw_score, 0), 100)
            weighted_sum += normalized_score * weights['weight']
            total_weight += weights['weight']

    # Si total_weight est 0, retourner 0
    if total_weight == 0:
        return 0

    return weighted_sum / total_weight

def _assign_rating(self, score):
    """Attribue une notation par √©toiles"""
    if score >= 85:
        return '‚òÖ‚òÖ‚òÖ‚òÖ‚òÖ (Excellent)'
    elif score >= 70:
        return '‚òÖ‚òÖ‚òÖ‚òÖ (Bon)'
    elif score >= 55:
        return '‚òÖ‚òÖ‚òÖ (Moyen)'
    elif score >= 40:
        return '‚òÖ‚òÖ (Faible)'
    else:
        return '‚òÖ (Tr√®s faible)'

def _assess_risk_level(self, e_score, s_score, g_score):
    """√âvalue le niveau de risque ESG"""
    avg_score = (e_score + s_score + g_score) / 3

    if avg_score >= 75:
        return 'Risque faible'
    elif avg_score >= 60:
        return 'Risque mod√©r√©'
    elif avg_score >= 45:
        return 'Risque √©lev√©'
    else:
        return 'Risque tr√®s √©lev√©'

def _categorize_performance(self, score):
    """Cat√©gorise la performance ESG"""
    if score >= 80:
        return 'Leader'
    elif score >= 65:
        return 'Avanc√©'
    elif score >= 50:
        return 'Moyen'
    elif score >= 35:
        return 'En retard'
    else:
        return 'Critique'

def _generate_recommendations(self, company, e_score, s_score, g_score):
    """G√©n√®re des recommandations personnalis√©es"""
    recommendations = []
    company_id = company['Company_ID']
    company_name = company['Company_Name']

    # Recommandations environnementales
    if e_score < 60:
        if company.get('GHG_Reduction', 100) < 50:
            recommendations.append({
                'Company_ID': company_id,
                'Company_Name': company_name,
                'Category': 'Environnement',
                'Priority': 'Haute',
                'Recommendation': 'Mettre en place un plan de r√©duction des √©missions CO2 avec objectifs quantifi√©s',
                'Norme': 'GRI 305, TCFD'
            })

    # Recommandations sociales
    if s_score < 60:
        if company.get('Diversity_Index', 100) < 40:
            recommendations.append({
                'Company_ID': company_id,
                'Company_Name': company_name,
                'Category': 'Social',
                'Priority': 'Moyenne',
                'Recommendation': 'Am√©liorer la diversit√© au sein du conseil et de la direction',
                'Norme': 'SDG 5, SDG 10'
            })

    # Recommandations de gouvernance
    if g_score < 60:
        if company.get('Transparency_Index', 100) < 50:
            recommendations.append({
                'Company_ID': company_id,
                'Company_Name': company_name,
                'Category': 'Gouvernance',
                'Priority': 'Haute',
                'Recommendation': 'Augmenter la transparence sur les pratiques ESG',
                'Norme': 'GRI 205'
            })

    return recommendations

# Ajouter les m√©thodes √† la classe
ESGPortfolioAnalyzer.calculate_detailed_scores = calculate_detailed_scores
ESGPortfolioAnalyzer._calculate_weighted_score = _calculate_weighted_score
ESGPortfolioAnalyzer._assign_rating = _assign_rating
ESGPortfolioAnalyzer._assess_risk_level = _assess_risk_level
ESGPortfolioAnalyzer._categorize_performance = _categorize_performance
ESGPortfolioAnalyzer._generate_recommendations = _generate_recommendations

print("‚úÖ Partie 5/6 : M√©thodes de calcul ajout√©es!")
```

```python
"""
Portfolio ESG Analyzer - Partie 6/6 : Calcul principal et visualisations
"""

def calculate_esg_scores(self):
    """Calcule les scores ESG pour chaque entreprise"""
    if self.portfolio_data is None:
        raise ValueError("‚ùå Veuillez d'abord charger les donn√©es du portefeuille")

    print("\nüßÆ CALCUL DES SCORES ESG COMPLETS")
    print("=" * 40)
    results = []

    for _, company in self.portfolio_data.iterrows():
        # Calcul des scores par cat√©gorie avec pond√©ration
        env_score = self._calculate_weighted_score(company, 'environmental')
        soc_score = self._calculate_weighted_score(company, 'social')
        gov_score = self._calculate_weighted_score(company, 'governance')

        # Score total pond√©r√©
        total_score = (
            env_score * self.esg_framework['environmental']['total_weight'] +
            soc_score * self.esg_framework['social']['total_weight'] +
            gov_score * self.esg_framework['governance']['total_weight']
        )

        # Classification par √©toiles
        rating = self._assign_rating(total_score)

        # Niveau de risque
        risk_level = self._assess_risk_level(env_score, soc_score, gov_score)

        # Cat√©gorie de performance
        performance_category = self._categorize_performance(total_score)

        # Recommandations
        recommendations = self._generate_recommendations(company, env_score, soc_score, gov_score)

        company_result = {
            'Company_ID': company['Company_ID'],
            'Company_Name': company['Company_Name'],
            'Sector': company['Sector'],
            'Country': company.get('Country', 'N/A'),
            'E_Score': round(env_score, 1),
            'S_Score': round(soc_score, 1),
            'G_Score': round(gov_score, 1),
            'Total_ESG_Score': round(total_score, 1),
            'ESG_Rating': rating,
            'Risk_Level': risk_level,
            'Performance_Category': performance_category,
            'Recommendations_Count': len(recommendations)
        }
        results.append(company_result)

        # Stocker les recommandations
        self.recommendations.extend(recommendations)

    self.results = pd.DataFrame(results)
    self._generate_insights()

    print(f"‚úÖ Scores calcul√©s pour {len(results)} entreprises")
    return self.results

def _generate_insights(self):
    """G√©n√®re des insights automatiques"""
    print("\nüí° G√âN√âRATION D'INSIGHTS")
    print("=" * 40)

    self.insights = {
        'portfolio_summary': {
            'total_companies': len(self.results),
            'avg_esg_score': round(self.results['Total_ESG_Score'].mean(), 1),
            'median_esg_score': round(self.results['Total_ESG_Score'].median(), 1),
            'std_esg_score': round(self.results['Total_ESG_Score'].std(), 1),
            'best_score': round(self.results['Total_ESG_Score'].max(), 1),
            'worst_score': round(self.results['Total_ESG_Score'].min(), 1)
        },
        'performance_distribution': {
            'leaders': len(self.results[self.results['Performance_Category'] == 'Leader']),
            'advanced': len(self.results[self.results['Performance_Category'] == 'Avanc√©']),
            'average': len(self.results[self.results['Performance_Category'] == 'Moyen']),
            'lagging': len(self.results[self.results['Performance_Category'] == 'En retard']),
            'critical': len(self.results[self.results['Performance_Category'] == 'Critique'])
        },
        'risk_analysis': {
            'low_risk': len(self.results[self.results['Risk_Level'] == 'Risque faible']),
            'moderate_risk': len(self.results[self.results['Risk_Level'] == 'Risque mod√©r√©']),
            'high_risk': len(self.results[self.results['Risk_Level'] == 'Risque √©lev√©']),
            'very_high_risk': len(self.results[self.results['Risk_Level'] == 'Risque tr√®s √©lev√©'])
        },
        'top_performers': self.results.nlargest(3, 'Total_ESG_Score')[['Company_Name', 'Sector', 'Total_ESG_Score', 'ESG_Rating']].to_dict('records'),
        'bottom_performers': self.results.nsmallest(3, 'Total_ESG_Score')[['Company_Name', 'Sector', 'Total_ESG_Score', 'ESG_Rating']].to_dict('records'),
        'sector_analysis': self.results.groupby('Sector').agg({
            'Total_ESG_Score': ['mean', 'median', 'count'],
            'Risk_Level': lambda x: x.mode()[0] if not x.mode().empty else 'N/A'
        }).round(1).to_dict()
    }

    # Calcul des piliers
    esg_pillars = {
        'avg_e_score': round(self.results['E_Score'].mean(), 1),
        'avg_s_score': round(self.results['S_Score'].mean(), 1),
        'avg_g_score': round(self.results['G_Score'].mean(), 1)
    }

    # D√©terminer le pilier le plus fort et le plus faible
    pillar_scores = {
        'E': esg_pillars['avg_e_score'],
        'S': esg_pillars['avg_s_score'],
        'G': esg_pillars['avg_g_score']
    }

    esg_pillars['strongest_pillar'] = max(pillar_scores, key=pillar_scores.get)
    esg_pillars['weakest_pillar'] = min(pillar_scores, key=pillar_scores.get)

    self.insights['esg_pillar_comparison'] = esg_pillars

    # Affichage des insights
    print("\nüìä INSIGHTS PRINCIPAUX:")
    print(f"‚Ä¢ Score ESG moyen du portefeuille: {self.insights['portfolio_summary']['avg_esg_score']}/100")
    print(f"‚Ä¢ Entreprises leaders: {self.insights['performance_distribution']['leaders']}")
    print(f"‚Ä¢ Entreprises √† risque √©lev√©: {self.insights['risk_analysis']['high_risk'] + self.insights['risk_analysis']['very_high_risk']}")
    print(f"‚Ä¢ Pilier le plus fort: {self.insights['esg_pillar_comparison']['strongest_pillar']}")
    print(f"‚Ä¢ Pilier le plus faible: {self.insights['esg_pillar_comparison']['weakest_pillar']}")

def generate_comprehensive_report(self):
    """G√©n√®re un rapport complet Excel avec plusieurs feuilles"""
    if self.results.empty:
        raise ValueError("‚ùå Veuillez d'abord calculer les scores ESG")

    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    output_file = f'ESG_Portfolio_Report_{timestamp}.xlsx'

    print(f"\nüìä G√©n√©ration du rapport Excel: {output_file}")

    try:
        with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
            # 1. R√©sum√© g√©n√©ral
            summary_df = self.results[['Company_Name', 'Sector', 'Country', 'ESG_Rating',
                                      'Risk_Level', 'Performance_Category', 'Total_ESG_Score']]
            summary_df.to_excel(writer, sheet_name='R√©sum√©', index=False)

            # 2. Scores d√©taill√©s
            self.results.to_excel(writer, sheet_name='Scores_D√©taill√©s', index=False)

            # 3. Scores par crit√®re
            if not self.detailed_scores.empty:
                self.detailed_scores.to_excel(writer, sheet_name='Scores_Crit√®res', index=False)

            # 4. Recommandations
            if self.recommendations:
                rec_df = pd.DataFrame(self.recommendations)
                rec_df.to_excel(writer, sheet_name='Recommandations', index=False)

            # 5. Insights et analyses
            insights_data = []
            for category, data in self.insights.items():
                if isinstance(data, dict):
                    for key, value in data.items():
                        insights_data.append({'Cat√©gorie': category, 'M√©trique': key, 'Valeur': str(value)})
                else:
                    insights_data.append({'Cat√©gorie': category, 'M√©trique': 'Valeur', 'Valeur': str(data)})

            pd.DataFrame(insights_data).to_excel(writer, sheet_name='Insights', index=False)

        print(f"‚úÖ Rapport Excel g√©n√©r√©: {output_file}")
        return output_file

    except Exception as e:
        print(f"‚ùå Erreur lors de la g√©n√©ration du rapport: {e}")
        return None

def generate_json_report(self):
    """G√©n√®re un rapport JSON complet"""
    if self.results.empty:
        raise ValueError("‚ùå Veuillez d'abord calculer les scores ESG")

    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    output_file = f'ESG_Portfolio_Report_{timestamp}.json'

    try:
        report_data = {
            'report_metadata': {
                'generated_at': timestamp,
                'portfolio_size': len(self.results),
                'analysis_date': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
                'esg_framework_version': '2.0'
            },
            'portfolio_summary': self.insights,
            'companies': self.results.to_dict('records'),
            'recommendations': self.recommendations
        }

        with open(output_file, 'w', encoding='utf-8') as f:
            json.dump(report_data, f, ensure_ascii=False, indent=2)

        print(f"‚úÖ Rapport JSON g√©n√©r√©: {output_file}")
        return output_file

    except Exception as e:
        print(f"‚ùå Erreur lors de la g√©n√©ration du rapport JSON: {e}")
        return None

def visualize_results(self, save_figures=True):
    """Cr√©e des visualisations pour le rapport"""
    if self.results.empty:
        raise ValueError("‚ùå Veuillez d'abord calculer les scores ESG")

    print("\nüé® G√âN√âRATION DES VISUALISATIONS")
    print("=" * 40)

    try:
        # Cr√©er une figure avec plusieurs sous-graphiques
        fig, axes = plt.subplots(2, 2, figsize=(15, 12))

        # 1. Distribution des scores ESG
        ax1 = axes[0, 0]
        sns.histplot(data=self.results, x='Total_ESG_Score', bins=10, kde=True, ax=ax1)
        ax1.axvline(self.results['Total_ESG_Score'].mean(), color='red', linestyle='--',
                   label=f'Moyenne: {self.results["Total_ESG_Score"].mean():.1f}')
        ax1.set_title('Distribution des Scores ESG', fontsize=12, fontweight='bold')
        ax1.set_xlabel('Score ESG')
        ax1.set_ylabel('Nombre d\'entreprises')
        ax1.legend()

        # 2. Scores par secteur
        ax2 = axes[0, 1]
        sector_scores = self.results.groupby('Sector')['Total_ESG_Score'].mean().sort_values()
        bars = ax2.barh(range(len(sector_scores)), sector_scores.values)
        ax2.set_yticks(range(len(sector_scores)))
        ax2.set_yticklabels(sector_scores.index)
        ax2.set_title('Score ESG Moyen par Secteur', fontsize=12, fontweight='bold')
        ax2.set_xlabel('Score ESG moyen')

        # Ajouter les valeurs sur les barres
        for i, (bar, score) in enumerate(zip(bars, sector_scores.values)):
            ax2.text(score + 0.5, i, f'{score:.1f}', va='center')

        # 3. Matrice de risque
        ax3 = axes[1, 0]
        risk_counts = self.results['Risk_Level'].value_counts()
        colors = ['#2ecc71', '#f39c12', '#e74c3c', '#c0392b']
        wedges, texts, autotexts = ax3.pie(risk_counts.values, labels=risk_counts.index,
                                          autopct='%1.1f%%', colors=colors, startangle=90)
        ax3.set_title('Distribution des Niveaux de Risque', fontsize=12, fontweight='bold')

        # 4. Comparaison des piliers ESG
        ax4 = axes[1, 1]
        esg_components = self.results[['E_Score', 'S_Score', 'G_Score']].mean()
        bars = ax4.bar(['Environnement', 'Social', 'Gouvernance'], esg_components.values,
                      color=['#27ae60', '#3498db', '#9b59b6'])
        ax4.set_title('Scores Moyens par Pilier ESG', fontsize=12, fontweight='bold')
        ax4.set_ylim(0, 100)
        ax4.set_ylabel('Score moyen')

        # Ajouter les valeurs sur les barres
        for bar, value in zip(bars, esg_components.values):
            height = bar.get_height()
            ax4.text(bar.get_x() + bar.get_width()/2., height + 1,
                    f'{value:.1f}', ha='center', va='bottom')

        plt.suptitle('Analyse ESG du Portefeuille', fontsize=14, fontweight='bold')
        plt.tight_layout()

        if save_figures:
            plt.savefig('ESG_Portfolio_Visualization.png', dpi=300, bbox_inches='tight')
            print("‚úÖ Visualisations sauvegard√©es: ESG_Portfolio_Visualization.png")

        plt.show()

    except Exception as e:
        print(f"‚ùå Erreur lors de la g√©n√©ration des visualisations: {e}")

# Ajouter les m√©thodes √† la classe
ESGPortfolioAnalyzer.calculate_esg_scores = calculate_esg_scores
ESGPortfolioAnalyzer._generate_insights = _generate_insights
ESGPortfolioAnalyzer.generate_comprehensive_report = generate_comprehensive_report
ESGPortfolioAnalyzer.generate_json_report = generate_json_report
ESGPortfolioAnalyzer.visualize_results = visualize_results

print("‚úÖ Partie 6/6 : Calcul principal et visualisations ajout√©s!")
```

```python
"""
Portfolio ESG Analyzer - Partie 7/7 : Fonction d'ex√©cution principale
"""

def run_complete_analysis():
    """Ex√©cute l'analyse compl√®te du portefeuille ESG"""
    print("=" * 70)
    print("üöÄ ESG PORTFOLIO ANALYZER - VERSION 2.0")
    print("=" * 70)

    # Initialiser l'analyseur
    analyzer = ESGPortfolioAnalyzer()

    # Chemin du fichier
    excel_file = 'portfolio_data.xlsx'

    # V√©rifier si le fichier existe
    if not os.path.exists(excel_file):
        print(f"‚ùå Fichier '{excel_file}' non trouv√©!")
        print("\nüìù Veuillez ex√©cuter la cellule de t√©l√©versement (Partie 2)")
        return

    print(f"\nüìÅ Fichier source: {excel_file}")
    print("=" * 70)

    # √âtape 1: Charger les donn√©es
    print("\nüì• √âTAPE 1: CHARGEMENT DES DONN√âES")
    print("-" * 40)
    if not analyzer.load_portfolio_data(excel_file):
        return

    # √âtape 2: Calculer les scores d√©taill√©s
    print("\nüßÆ √âTAPE 2: CALCUL DES SCORES D√âTAILL√âS")
    print("-" * 40)
    analyzer.calculate_detailed_scores()

    # √âtape 3: Calculer les scores ESG
    print("\nüìä √âTAPE 3: CALCUL DES SCORES ESG")
    print("-" * 40)
    results = analyzer.calculate_esg_scores()

    # Afficher un aper√ßu des r√©sultats
    print("\nüìã APER√áU DES R√âSULTATS:")
    print(results[['Company_Name', 'Sector', 'Total_ESG_Score', 'ESG_Rating', 'Risk_Level']].head())

    # √âtape 4: G√©n√©rer les rapports
    print("\nüìÑ √âTAPE 4: G√âN√âRATION DES RAPPORTS")
    print("-" * 40)
    excel_report = analyzer.generate_comprehensive_report()
    json_report = analyzer.generate_json_report()

    # √âtape 5: Visualisations
    print("\nüé® √âTAPE 5: VISUALISATIONS")
    print("-" * 40)
    analyzer.visualize_results(save_figures=True)

    # √âtape 6: R√©sultats finaux
    print("\n" + "=" * 70)
    print("üèÜ R√âSULTATS FINAUX")
    print("=" * 70)

    # Afficher le top 5
    print("\nüèÜ TOP 5 DES ENTREPRISES:")
    top_5 = results.nlargest(5, 'Total_ESG_Score')
    for idx, (_, row) in enumerate(top_5.iterrows(), 1):
        print(f"{idx}. {row['Company_Name']} ({row['Sector']})")
        print(f"   Score: {row['Total_ESG_Score']} | Rating: {row['ESG_Rating']} | Risque: {row['Risk_Level']}")
        print()

    # Afficher les recommandations principales
    if analyzer.recommendations:
        print("\nüîß PRINCIPALES RECOMMANDATIONS:")
        high_priority_recs = [r for r in analyzer.recommendations if r['Priority'] == 'Haute']
        for idx, rec in enumerate(high_priority_recs[:3], 1):
            print(f"{idx}. {rec['Company_Name']} - {rec['Category']}")
            print(f"   {rec['Recommendation']}")
            print(f"   Norme: {rec['Norme']}")
            print()

    # R√©sum√©
    print("\n" + "=" * 70)
    print("‚úÖ ANALYSE TERMIN√âE AVEC SUCC√àS!")
    print("=" * 70)

    # Fichiers g√©n√©r√©s
    print("\nüìÅ FICHIERS G√âN√âR√âS:")
    if excel_report:
        print(f"1. {excel_report} - Rapport Excel complet")
    if json_report:
        print(f"2. {json_report} - Rapport JSON structur√©")
    print("3. ESG_Portfolio_Visualization.png - Visualisations")

    # Sauvegarder √©galement en CSV
    results.to_csv('esg_results_summary.csv', index=False, encoding='utf-8-sig')
    print("4. esg_results_summary.csv - R√©sum√© au format CSV")

    print("\nüéØ PROCHAINES √âTAPES:")
    print("‚Ä¢ Examiner les recommandations sp√©cifiques par entreprise")
    print("‚Ä¢ Comparer les performances sectorielles")
    print("‚Ä¢ Planifier des actions d'am√©lioration ESG")

    return analyzer, results

print("‚úÖ Partie 7/7 : Fonction d'ex√©cution principale ajout√©e!")
print("\nüéâ TOUTES LES PARTIES SONT PR√äTES!")
print("\nüìã POUR D√âMARRER L'ANALYSE:")
print("1. Ex√©cutez toutes les cellules dans l'ordre (1 √† 7)")
print("2. T√©l√©versez votre fichier Excel quand demand√©")
print("3. Ex√©cutez: analyzer, results = run_complete_analysis()")
```
